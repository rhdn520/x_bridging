```python
import json

with open("gpt_intps.json", 'r') as f:
    gpt_data = json.load(f)

with open("diffusion_intps.json", 'r') as f:
    diff_data = json.load(f)
```

```python
from sacrebleu.metrics import BLEU, CHRF

chrf = CHRF()
```

```python
chrf_scores = []
delta_scores = []

for diff_intp, gpt_intp in zip(diff_data, gpt_data):
    # diff_intp[1]
    score_1 = chrf.sentence_score(diff_intp[1], [gpt_intp[1]]).score
    score_2 = chrf.sentence_score(diff_intp[2], [gpt_intp[2]]).score
    score_3 = chrf.sentence_score(diff_intp[3], [gpt_intp[3]]).score
    
    
    score_corpus = chrf.corpus_score(
                diff_intp[1:4],
                [[x] for x in gpt_intp[1:4]]
            ).score
    
    chrf_scores.append({
        'score_1': score_1,
        'score_2': score_2,
        'score_3': score_3,
        'corpus_score': score_corpus
    })

    delta_values = []
    for sent in diff_intp:
        delta_values.append(chrf.sentence_score(sent, [diff_intp[-1]]).score)

    delta_scores.append(delta_values)
```

```python
analysis_result = {
    'chrf_scores': chrf_scores,
    'delta_scores': delta_scores
}
```

```python
with open('analysis_result.json', 'w', encoding='utf-8') as f:
    json.dump(analysis_result, f, ensure_ascii=False, indent=4)
```
