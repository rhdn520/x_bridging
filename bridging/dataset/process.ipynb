```python
import json

ftext = "train"
with open(f"/home/seungwoochoi/data/x_bridging/bridging/dataset/{ftext}.jsonl") as f:
    # print(f.readlines())
    texts = []
    for line in f.readlines():
        tmp = json.loads(line)
        texts.append(tmp['src'].strip())
        texts.append(tmp['trg'].strip())
    
with open(f"/home/seungwoochoi/data/x_bridging/bridging/dataset/{ftext}.txt", 'w') as f:
    f.write("\n".join(texts))
```
